```python
#Convert target to binary (1=good, 2=bad) -> (0=good, 1=bad)
y = y.map({1: 0, 2: 1})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print("X_train type:", type(X_train))
print("y_train type:", type(y_train))
print("X_train shape:", getattr(X_train, "shape", "No shape"))
print("y_train shape:", getattr(y_train, "shape", "No shape"))
print("First few values of y_train:", y_train[:5])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier(max_depth=5, random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
plt.figure(figsize=(20,10))
plot_tree(clf, feature_names=X.columns, class_names=['Good', 'Bad'], filled=True)
plt.title("Decision Tree for German Credit Data")
plt.show()
```
